### Brute Force Search: Find Two Numbers That Sum to a Target

This example demonstrates a brute-force approach to find if any two numbers in a given list add up to a specific target sum. It checks every possible pair of numbers.

```python
def find_pair_with_sum_brute_force(numbers, target_sum):
    n = len(numbers)
    for i in range(n):
        for j in range(i + 1, n):
            if numbers[i] + numbers[j] == target_sum:
                return (numbers[i], numbers[j])
    return None

# Example usage:
my_list = [2, 7, 11, 15, 3, 6, 8]
target = 9

print(f"List: {my_list}")
print(f"Target Sum: {target}")

result = find_pair_with_sum_brute_force(my_list, target)

if result:
    print(f"Found pair: {result[0]} and {result[1]} sum up to {target}")
else:
    print(f"No pair found that sums up to {target}")

target_not_found = 20
print(f"\nTarget Sum: {target_not_found}")
result_not_found = find_pair_with_sum_brute_force(my_list, target_not_found)
if result_not_found:
    print(f"Found pair: {result_not_found[0]} and {result_not_found[1]} sum up to {target_not_found}")
else:
    print(f"No pair found that sums up to {target_not_found}")
```
