```python
def bubble_sort(arr):
    length = len(arr)

    for i in range(0, length - 1):
        for j in range(0, length - i - 1):

            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

    return arr


if __name__ == "__main__":
    arr = list(map(int, input().split(',')))
    sort = bubble_sort(arr)

    print(sort)
```
