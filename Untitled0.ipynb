```python
class Student:

    """Docstring of the class"""

    def __init__(self):
        # instance variables
        self.name = "Anurag"
        self.roll_no = 101
        self.marks=500

    def display(self):
        # instance method
        print("The name is:", self.name)
        print("The roll number is:", self.roll_no)
        print("The marks is:", self.marks)


# object creation
s1 = Student()

# calling method
s1.display()
```

```python
class Student:
    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def display(self):
        print(f"The name is: {self.name}")
        print(f"The roll number is: {self.roll_no}")
        print(f"The marks is: {self.marks}")

# object creation
s2 = Student("Alice", 102, 550)

# calling method
s2.display()
```

```python
class Value:
  def __init__(self):
    self.x=10
  def modify(self):
    self.x=self.x+1
  def display(self):
    print(self.x)
```

```python

v=Value()
v2=Value()
print(v.display()) #10
v.modify()
print(v.display()) #11
print(v2.display()) #10
```

```python
class Sample:
  x = 10 # class variable or static variable

  @classmethod
  def modify(cls):
    # This method modifies the class variable 'x'
    cls.x += 1

  @classmethod
  def display(cls):
    print(cls.x)

# Initial display of x
Sample.display()

# Display x again to match desired output
Sample.display()

# Modify x using the class method
Sample.modify()

# Display x after modification
Sample.display()

# Another modification
Sample.modify()
Sample.display()
```
