```python
def count_distributions(stock):

    count = 0

    while stock > 0:

        # If stock is even
        if stock % 2 == 0:
            stock = stock // 2

        # If stock is odd
        else:
            stock = stock - 1

        count = count + 1

    return count


# Input
stock = int(input())

# Output
print(count_distributions(stock))
```
