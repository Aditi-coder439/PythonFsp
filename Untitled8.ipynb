```python
def has_pair_with_sum(arr, target_sum):
    seen = set()

    for num in arr:
        complement = target_sum - num

        if complement in seen:
            return True

        seen.add(num)

    return False


# Test cases
arr1 = [1, 2, 4, 5]
target1 = 6

arr2 = [1, 2, 4, 5]
target2 = 10

print(has_pair_with_sum(arr1, target1))  # True
print(has_pair_with_sum(arr2, target2))  # False
```

```python
from typing import List

def max_profit(prices: List[int]) -> int:
    if not prices or len(prices) < 2:
        return 0

    min_price = float('inf')
    max_overall_profit = 0

    for price in prices:
        min_price = min(min_price, price)
        current_profit = price - min_price
        max_overall_profit = max(max_overall_profit, current_profit)

    return max_overall_profit

prices1 = [7, 1, 5, 3, 6, 4]
print(f"Prices: {prices1}, Max Profit: {max_profit(prices1)}")

prices2 = [7, 6, 4, 3, 1]
print(f"Prices: {prices2}, Max Profit: {max_profit(prices2)}")

prices3 = [2, 4, 1]
print(f"Prices: {prices3}, Max Profit: {max_profit(prices3)}")

prices4 = []
print(f"Prices: {prices4}, Max Profit: {max_profit(prices4)}")

prices5 = [1]
print(f"Prices: {prices5}, Max Profit: {max_profit(prices5)}")
```

```python
s = input("Enter input: ")

items = s.split(",")

result = ""

for item in items:
    name = ""
    digits = []

    for ch in item:
        if ch.isalpha():
            name += ch
        elif ch.isdigit():
            digits.append(int(ch))

    length = len(name)

    valid_digits = [d for d in digits if d <= length]

    if valid_digits:
        pos = max(valid_digits)
        result += name[pos - 1]
    else:
        result += "X"

print(result)
```

```python
s = input("Enter a string: ")

chars = list(s)

left = 0
right = len(chars) - 1

while left < right:

    if not chars[left].isalpha():
        left += 1

    elif not chars[right].isalpha():
        right -= 1

    else:
        chars[left], chars[right] = chars[right], chars[left]
        left += 1
        right -= 1

print("Output:", "".join(chars))
```

```python
arr = list(map(int, input().split()))

start = arr.index(4)
end = arr.index(7)

number1 = 0
middle = ""

for i in range(len(arr)):
    if i > start and i < end:
        middle += str(arr[i])
    elif i < start or i > end:
        number1 += arr[i]

number2 = int(middle)

print(number1 + number2)
```
